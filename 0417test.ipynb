```python
from __future__ import print_function, division

%matplotlib inline

import thinkdsp
import thinkplot
import numpy as np
import math

import warnings
warnings.filterwarnings('ignore')

from IPython.html.widgets import interact, interact_manual, fixed
from IPython.html import widgets
from IPython.display import display

PI2 = 2 * math.pi
```

```python
from thinkdsp import Sinusoid
import numpy as np

class chuSawtoothSignal(thinkdsp.Sinusoid):
        
    def evaluate(self, ts):
    
        θ= ts * 2 * np.pi * self.freq + self.offset # 在此作橫軸之變數變換

        θ= θ % (np.pi * 2) # 這一行讓 θ 周而復始，週期維持在 (2 π)

        x= -1 + θ/np.pi # 從 -1 開始爬坡，上半週期到達　0,下半周期到達 +1 
       
        x= x * self.amp # 最後輸出前，再把縱軸拉到 振幅 = amp

        return x
```

```python
import matplotlib.pylab as pl

x= chuSawtoothSignal(freq=1000,amp=10)
x.plot()
pl.grid()
```

```python
w= x.make_wave()
w.segment(duration=1000).plot()
w.make_audio()
```
